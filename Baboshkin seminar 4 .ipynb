17) Отсортировать вектор из 10 случайных элементов.

```python
rand_vec = np.random.random(10)
print(rand_vec)
print(sorted(rand_vec))
for j in range(len(rand_vec) - 1):
    f = 0
    for i in range(len(rand_vec) - 1 - j):
        prev = rand_vec[i]
        nxt = rand_vec[i+1]
        if prev > nxt:
            rand_vec[i+1] = prev
            rand_vec[i] = nxt
            f = 1
        if f == 0:
            break
print(rand_vec)
```

18) Найти ближайшее к заданному значению число в заданном массиве.

```python
wanted = 12
rand_vc = 100 * np.random.random(10)
closest = rand_vc[np.argmin(abs(rand_vc - wanted))]
print(sorted(rand_vc), closest)
```

19) Отнять среднее из каждой строки в матрице.

```python
rm = 20 * np.random.random(20).reshape(4,5)
norms = rm.copy()
print(norms)
for i in range(len(norms)):
    norms[i] -= np.mean(rm[i])
print(norms)  
```

20) Дан четырехмерный массив, посчитать сумму по последним двум осям.
