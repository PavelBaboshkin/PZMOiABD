20. Создать датафрейм в котором есть столбец, содержащий объем торгов (столбец 7),
нормированный по среднему значению объема торгов для данного тикера.

```python
mean_vol = sp.groupby('ticker').mean().volume.copy()
vol_merged = sp.merge(mean_vol, left_on='ticker', right_on='ticker')
vol_merged['volume_normalized'] = vol_merged.volume_x/vol_merged.volume_y
vol_merged.groupby('ticker').mean()
```
