1) Постройте график линейной функции (kx+b) с коэффициентами k и b, вводимыми
пользователем. Предварительно рассчитать 100 значений аргумента из диапазона
введенного пользователем.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

coeffs = [int(i) for i in input('Введите коэффициенты B и K и границы диапазона X: ').split()]
x = np.linspace(coeffs[2], coeffs[3], 100)
y = x*coeffs[1] + coeffs[0]
plt.title(f'Линейная зависимость y = {coeffs[0]} + {coeffs[1]} * x',  color='w')
plt.grid()
plt.plot(x,y)
```

2. Постройте графики y = sin(x) и y = e -0,1x cos(x) на одной картинке. Сделайте их
разными цветами и добавьте подписи.

```python
start, end = -20, 10
x2 = np.linspace(start,end,1000)
y2_1 = np.sin(x2)
y2_2 = np.exp(-0.1*x2)*np.cos(x2)
plt.title(f'Графики y = sin(x) и y = e -0,1x cos(x)',  color='w')
plt.xlabel('Ось X', color = 'w') 
plt.ylabel('Ось Y1, Y2' , color = 'w')
plt.grid()
plt.plot(x2,y2_1,label = 'y = sin(x)', color = 'r')
plt.plot(x2,y2_2,label = 'y = e^(-0,1x) cos(x)', color = 'g')
plt.legend()
```

3. Постройте графики y = x 2 и y = |x| на двух картинках в строке. Добавьте легенды.

```python
start, end = -5, 5
x3 = np.linspace(start,end,100)
y3_1 = x3 ** 2
y3_2 = abs(x3)

plt.figure(figsize=(12, 7))
plt.subplot(2, 2, 1)
plt.plot(x3, y3_1, label = 'y = x**2', color = 'g')
plt.legend()
plt.subplot(2, 2, 2)
plt.plot(x3, y3_2, label = 'y = |x|', color = 'g')
plt.legend()
```

4. Сгенерируйте 1000 случайных величин с равномерным распределение от 0 до 10 и
постройте гистограмму.

```python
x = np.random.random(1000) * 10
n, bins, patches = plt.hist(x, 10, density=True, facecolor='g', alpha=0.75)

plt.xlabel('value', color = 'w')
plt.ylabel('Probability', color = 'w')
plt.title('Histogram of distribution', color = 'w')
plt.grid()
plt.show()
```

5. Сгенерируйте 1000 случайных величин с нормальным распределением (нулевое
среднее и дисперсия равна 5). Постройте гистограмму.

```python
mu, sigma = 0, 5
x = np.random.normal(mu, sigma, 1000)
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)

plt.xlabel('x', color = 'w')
plt.ylabel('Probability', color = 'w')
plt.title('Histogram of normal distribution', color = 'w')
plt.text(-14, .07, f'mu = {mu}, sigma = {sigma}')
plt.grid()
plt.show()
```

6. Сгенерируйте случайный процесс X(t) = r*X(t-1) + e(t), где дисперсия случайной
нормальной добавки e = (1-r^2)*Dx. Отобразите его на графике.

```python
r = 0.1
Dx = 1
De = (1-r**2)*Dx
x = np.zeros(200)
x[0] = Dx*np.random.randn()
e = De*np.random.randn(200)
for i in range(1,200):
    x[i] = x[i-1]*r + e[i]
plt.plot(x)
```

```python
plt.xlabel('Xk+1', color = 'w')
plt.ylabel('Xk', color = 'w')
plt.title('Scatter of Xk by Xk+1', color = 'w')
plt.scatter(x[:-1],x[1:])
```

extra) Xk = 2rX(k-1) - r**2 X(k-2) + ek (дисперсия De = Dx(1-r**2)/(1+r**2)) 
построить 2 диаграммы рассеивания Xk Xk+1
Xk Xk+2

```python
r = 0.7
Dx = 1
De = Dx*((1-r**2)/(1+r**2))
x = np.zeros(200)
x[0] = Dx*np.random.randn()
e = De*np.random.randn(200)
for i in range(1,200):
    x[i] = 2*r*x[i-1]- r**2*x[i-2] + e[i]
plt.plot(x)
```

```python
plt.figure(figsize=(12, 7))
plt.subplot(2, 2, 1)
plt.scatter(x[:-1], x[1:], label = 'Xk, Xk+1', color = 'g')
plt.legend()
plt.subplot(2, 2, 2)
plt.scatter(x[:-2], x[2:], label = 'Xk, Xk+2', color = 'g')
plt.legend()
```
