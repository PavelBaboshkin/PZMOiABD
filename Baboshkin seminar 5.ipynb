x * sin(x) больше 0.5.

```python
approximation = 3
base_line = 0.5
start, end = 0, 100

x = np.linspace(start, end, 10 ** approximation)
y = x * np.sin(x)

xbool = y > base_line
y1 = x[xbool]
adding = x.copy()
adding[:] = np.nan
for i in range(len(xbool)):
    if xbool[i] == True:
        adding[i] = 0.5

import matplotlib.pyplot as plt
plt.plot(x,y)
plt.plot(x, adding)
```

6) Написать функцию, которая для целых значений n и m будет возвращать матрицу
размера n на m, имеющую заполнение вида. Пример для 4 на 4.


```python
n,m = 7,9
def like_matrix(n,m):
    first = [0,1]
    second = [2,3]
    matrix = []
    f = [i*first[:] for i in range(1,m//2+1)]
    s = [i*second[:] for i in range(1,m//2+1)]
    if m % 2 != 0:
        f[-1].append(0)
        s[-1].append(2)
    for i in range(n+1):
        if i % 2 == 0:
            matrix.append(f[-1])
        else:
            matrix.append(s[-1])
    return np.matrix(matrix)
print(like_matrix(n,m))
```

7) Написать функцию поиска всех простых чисел в пределах ста.

```python
def prime_numbers(upper):
    limitation = upper
    primes = []
    for i in range(2, limitation + 1):
        for j in primes:
            if i % j == 0:
                break
        else:
            primes.append(i)
    return primes
print(len(prime_numbers(100)))
```

9) При помощи прихотливого индексирования для двухмерного массива размерности
(20, 20), состоящего из случайных целых чисел в пределах от 0 до 10 получить
массив элементов находящихся на диагонали, проходящей над основной
диагональю.

```python
ndim = 24
data =np.random.randint(0,11,(ndim,ndim))
row,col = [],[]
for i in range(data.shape[1] - 1):
    row.append(i)
    col.append(i+1)
print(data)
print(data[row,col])
```
